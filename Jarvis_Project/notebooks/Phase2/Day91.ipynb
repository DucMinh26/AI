```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate

print("\n--- NGÀY 91: KHỞI ĐỘNG LANGCHAIN VÀ PROMPT TEMPLATE ---")

#1.API KEY
load_dotenv()
os.environ['GOOGLE_API_KEY'] = os.getenv('GEMINI_API_KEY')

#2.KHỞI TẠO LLM
llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash', temperature=0.7 )

#3.KHỞI TẠO PROMPT TEMPLATE
template_prompt = """
Bạn là một chuyên gia tóm tắt sách siêu đẳng.
Hãy tóm tắt ngắn gọn cuốn sách có tên là "{ten_sach}" trong đúng 3 gạch đầu dòng.
"""
prompt = PromptTemplate(input_variables=['ten_sach'], template=template_prompt)

#4.TẠO DÂY CHUYỀN (CHAIN) - Sức mạnh cốt lõi của LangChain
day_chuyen = prompt | llm

#5.Chạy thử dây chuyền
sach_can_hoi = 'Thiên tài bên trái, kẻ điên bên phải'
print(f"\n[HỆ THỐNG]: Đang yêu cầu tóm tắt sách '{sach_can_hoi}'...")

ket_qua = day_chuyen.invoke({'ten_sach':sach_can_hoi})

print("\n=> [AI TRẢ LỜI]:")
print("--------------------------------------------------")
# LangChain trả về một Object (AIMessage), ta cần gọi .content để lấy chữ
print(ket_qua.content)
print("--------------------------------------------------")
```
