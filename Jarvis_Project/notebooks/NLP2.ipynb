```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import time

print("\n--- NGÀY 72: HUẤN LUYỆN LSTM VỚI 25.000 ĐÁNH GIÁ PHIM IMDB ---")

#1.Tải dữ liệu IMDB(đã được mã hóa thành số sẵn)
#num_words = 10000 chỉ lấy 10.000 từ phổ biến nhất
print("Đang tải 50.000 bài đánh giá phim (MiniByte)... (Có thể mất vài giây)")
(X_train, y_train) ,(X_test, y_test) = imdb.load_data(num_words=10000)

print(f"Số lượng bài đánh giá phim trong tập huấn luyện: {len(X_train)}")
print(f"Số lượng bài đánh giá phim trong tập kiểm tra: {len(X_test)}")

#2.Tiền xử lí dữ liệu
#Số lượng từ trong mỗi bài đánh giá là khác nhau
#Ép độ dài tất cả đánh giá về 200 từ
X_train_pad = pad_sequences(X_train, maxlen=200)
X_test_pad = pad_sequences(X_test, maxlen=200)

#3.Xây dựng bộ não LSTM
model_lstm_imdb = Sequential()

#Lớp nhúng từ (Embedding layer)
model_lstm_imdb.add(Embedding(input_dim=10000, output_dim=32, input_length=200))

# Lớp Trí nhớ LSTM (64 nơ-ron: đủ mạnh để nhớ đoạn văn dài 200 chữ)
model_lstm_imdb.add(LSTM(64))

#Lớp đầu ra
model_lstm_imdb.add(Dense(1,activation='sigmoid'))

#4.Biên dịch mô hình
model_lstm_imdb.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

#5.Huấn luyện mô hình
print("\nĐang huấn luyện mô hình LSTM trên tập IMDB... (Có thể mất vài phút)")
start_time = time.time()

lich_su_hoc = model_lstm_imdb.fit(
    X_train_pad,
    y_train,
    epochs=3,
    batch_size=128,
    verbose=1
)
end_time=time.time()
print(f"Thời gian huấn luyện: {end_time - start_time:.2f} giây")

#6.Cho AI thi thử
print("\n--- AI LÀM BÀI THI ---")

loss, acc = model_lstm_imdb.evaluate(
    X_test_pad, 
    y_test, 
    batch_size=128,
    verbose=1
)

print(f"Độ chính xác trên tập kiểm tra: {acc*100:.2f}")
```
