```python
#Tắt bớt thông báo không cần thiết
import os 
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

from transformers import pipeline
import torch

print("\n--- NGÀY 77: ĐÁNH THỨC SỨC MẠNH TRANSFORMER ---")

#1. Khởi tạo đường ống (pipeline) sinh văn bản 
#pipeline() tự động tải mô hình GPT-2
#'text-generation','sentiment-analysis','translation': sinh chữ, phân tích cảm xúc, dịch thuật
print("Đang kết nối tới máy chủ Hugging Face và tải bộ não GPT-2...")
may_sinh_chu = pipeline('text-generation', model='gpt2')

#2. Đưa đề bài
cau_mo_dau = "In the future, Artificial Intelligence will"
print(f"\n[Câu hỏi]: {cau_mo_dau}")

#3. Sinh văn bản
print("\n[GPT-2 Đang suy nghĩ và viết tiếp...]")
ket_qua = may_sinh_chu(
    cau_mo_dau,
    max_length=50,
    num_return_sequences=1, #Trả về 1 đáp án
    temperature=0.7,
    pad_token_id=50256 #ký tự kết thúc, lấp đầy length của câu sẽ là 50256
)

#4. In kết quả
tac_pham = ket_qua[0]['generated_text']
print("\n[GPT-2 Trả lời]:")
print("--------------------------------------------------")
print(tac_pham)
print("--------------------------------------------------")
```

```python
print("\n--- NGÀY 78: KỸ NGHỆ PROMPT ENGINEERING ---")

prompt = """
Classify the sentiment of the following reviews as Positive or Negative.

Review: I love this product! It's amazing.
Sentiment: Positive

Review: This is the worst movie I have ever seen.
Sentiment: Negative

Review: The food was okay, but the service was terrible.
Sentiment: """

print(f"\n[Prompt]: {prompt}")

ket_qua = may_sinh_chu(
    prompt,
    max_new_tokens=2,
    num_return_sequences=1,
    temperature=0.1,
    pad_token_id=50256
)

cau_tra_loi_full = ket_qua[0]['generated_text'] #Lấy phần chữ mà AI sinh ra
cau_tra_loi = cau_tra_loi_full.split("Sentiment:")[-1].strip() #Chỉ lấy phần sau "Sentiment:" và loại bỏ khoảng trắng
print("\n=> [AI CHỐT ĐÁP ÁN]:", cau_tra_loi)
```

```python
from transformers import pipeline
import os
import torch

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

print("\n--- NGÀY 79: SỰ TIẾN HÓA THÀNH TRỢ LÝ ẢO (INSTRUCT MODEL) ---")

# 1. Tải mô hình Trợ lý ảo FLAN-T5 (Google)
# Lưu ý: Pipeline của T5 là 'text2text-generation' (Hỏi - Đáp) chứ không phải 'text-generation' (Viết tiếp)
print("Đang tải bộ não Trợ lý ảo FLAN-T5-Small... (Sẽ mất chút thời gian tải lần đầu)")
tro_ly_ai = pipeline('text2text-generation', model='google/flan-t5-small')

#2. Ra trực tiếp chứ không cần mớm prompt nữa
prompt = "Classify the sentiment of this review as Positive or Negative: The food was okay, but the service was terrible."
print(f"\n[Đề bài]: {prompt}")

#3. Trợ lý ảo trả lời
print("\n[Trợ lý AI đang làm việc...]")
ket_qua = tro_ly_ai(
    prompt,
    max_new_tokens=5
)

#4. In kết quả
ket_qua = ket_qua[0]['generated_text'].strip() #Lấy phần chữ mà AI sinh ra và loại bỏ khoảng trắng
print("\n=> [Trợ lý AI trả lời]:", ket_qua)

prompt2 ="Answer this question: What is the capital of Vietnam and what is its famous food?"
print(f"\n[Đề bài]: {prompt2}")
ket_qua2 = tro_ly_ai(
    prompt2,
    max_new_tokens=20
)
ket_qua2 = ket_qua2[0]['generated_text'].strip()
print("\n=> [Trợ lý AI trả lời]:", ket_qua2)
```
