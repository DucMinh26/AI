```python
import google.generativeai as genai
import os
from dotenv import load_dotenv

print("\n--- NGÀY 81: CHẾ TẠO CHATBOT CÓ TRÍ NHỚ ---")

#1.Bảo mật API
load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")
genai.configure(api_key=api_key)

#2. Khởi tạo mô hình
model = genai.GenerativeModel("gemini-2.5-flash")

#3.Kích hoạt phiên chat có trí nhớ
chat_session = model.start_chat(history=[])

print("Chào mừng đến với Trợ lý ảo Terminal! (Gõ 'thoát' hoặc 'exit' để dừng)\n")

#4.Vòng lặp Chat vô hạn

while True:
    user_input = input("User: ")

    if user_input.lower().strip() in ['thoát','exit','quit']:
        print("GoodBye!!!")
        break

    if user_input.strip() == "":
        continue

    try:
        #Gửi tin nhắn kèm lịch sử chat
        response = chat_session.send_message(user_input)
        print(f"Gemini: {response.text}\n")
    except Exception as e:
        print(f"[LỖI KẾT NỐI]: {e}\n")
```
