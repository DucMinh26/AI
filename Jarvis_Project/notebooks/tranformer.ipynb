```python
#Tắt bớt thông báo không cần thiết
import os 
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

from transformers import pipeline
import torch

print("\n--- NGÀY 77: ĐÁNH THỨC SỨC MẠNH TRANSFORMER ---")

#1. Khởi tạo đường ống (pipeline) sinh văn bản 
#pipeline() tự động tải mô hình GPT-2
#'text-generation','sentiment-analysis','translation': sinh chữ, phân tích cảm xúc, dịch thuật
print("Đang kết nối tới máy chủ Hugging Face và tải bộ não GPT-2...")
may_sinh_chu = pipeline('text-generation', model='gpt2')

#2. Đưa đề bài
cau_mo_dau = "In the future, Artificial Intelligence will"
print(f"\n[Câu hỏi]: {cau_mo_dau}")

#3. Sinh văn bản
print("\n[GPT-2 Đang suy nghĩ và viết tiếp...]")
ket_qua = may_sinh_chu(
    cau_mo_dau,
    max_length=50,
    num_return_sequences=1, #Trả về 1 đáp án
    temperature=0.7,
    pad_token_id=50256 #ký tự kết thúc, lấp đầy length của câu sẽ là 50256
)

#4. In kết quả
tac_pham = ket_qua[0]['generated_text']
print("\n[GPT-2 Trả lời]:")
print("--------------------------------------------------")
print(tac_pham)
print("--------------------------------------------------")
```
