```python
import google.generativeai as genai
import os
from dotenv import load_dotenv
print("\n--- NGÀY 80: MƯỢN BỘ NÃO KHỔNG LỒ TỪ ĐÁM MÂY (API) ---")

#1.Nạp biến môi trường từ file .env
load_dotenv()

#Lấy API key từ biến môi trường không hiện mã lên màn hình
api_key = os.getenv("GEMINI_API_KEY")
genai.configure(api_key=api_key)

#2.Lựa chọn mô hình
model = genai.GenerativeModel("gemini-2.5-flash")

#3.Câu hỏi
prompt ="What is the capital of Vietnam and what is its famous food?"

print(f"\n[CÂU HỎI]: {prompt}")
print("\n[Đang kết nối tới siêu máy tính của Google...]")

#4.Gửi câu hỏi và nhận câu trả lời
response = model.generate_content(prompt)

print("\n=> [GEMINI TRẢ LỜI]:")
print("--------------------------------------------------")
print(response.text)
print("--------------------------------------------------")
```
